# PyCaret

```python
from pycaret.classification import *

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = setup(data=train, 
              target='occyp_type', 
              train_size=0.8, # default value
              session_id=9,
              numeric_features=['income_total','DAYS_EMPLOYED'],
              categorical_features=['gender','car','reality','income_type','edu_type','family_type','house_type','work_phone','phone','email'])
```

```python
top_3_models = compare_models(sort='Accuracy', 
                              n_select = 3)
```

```python
evaluate_model(top_3_models[0]) #Random Forest Classifier
```

```python
model_top = top_3_models[0]
plot_model(model_top, plot = 'auc')
```

```python
predict_model(top_3_models[0],data=test)
```
